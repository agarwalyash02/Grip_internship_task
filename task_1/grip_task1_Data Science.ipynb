```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.txt',sep=",")
```

```python
df.head()
```

```python
plt.style.use('seaborn')
plt.scatter(df['Hours'],df['Scores'],marker="*",color="green")
plt.xlabel("Number of hours")
plt.ylabel("Score in percentage")
plt.title("Score of student over hours of study")
plt.show()
```

## Preparing the data

```python
from sklearn.model_selection import train_test_split
x = df['Hours'].values
y = df['Scores'].values
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
x_train = np.expand_dims(x_train,axis=1)
x_test = np.expand_dims(x_test,axis=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

## Training the algorithm

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(x_train,y_train)
print("Model trained")
```

```python
print(model.coef_[0],model.intercept_)
```

```python
regression_line = model.coef_[0]*x + model.intercept_
plt.style.use('seaborn')
plt.scatter(x,y,marker="*")
plt.plot(x,regression_line,color="green",marker='o',label="predicted")
plt.xlabel("Number of hours")
plt.ylabel("Score in percentage")
plt.title("Score of student over hours of study")
plt.legend()
plt.show()
```

## Making Predictions & Evaluating model

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics
print("The mean absolute error is :{}".format(metrics.mean_absolute_error(y_test,y_pred)))
print("The Root mean squared error is :{}".format(np.sqrt(metrics.mean_squared_error(y_test,y_pred))))
```

## Making prediction of the task query

```python
student_studied=9.25
predicted_score=model.predict([[student_studied]])
print("No of Hours = {}".format(student_studied))
print("Predicted Score = {}".format(predicted_score[0]))
```
